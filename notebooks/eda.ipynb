# EDA - Proyecto Final ML1

Este notebook contiene el análisis exploratorio de datos de las fuentes integradas.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraciones de visualización
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Carga de datos

```
df = pd.read_parquet('../data/final/features.parquet')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

## 2. Distribución del Nivel de Escalada (Target)

```
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='escalation_level', hue='country', palette='Set2')
plt.title('Distribución del Nivel de Escalada por País')
plt.xlabel('Nivel de Escalada (0=Bajo, 1=Medio, 2=Alto)')
plt.ylabel('Días')
plt.show()
```

## 3. Serie temporal de Tono Goldstein y Eventos

```
fig, axes = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

sns.lineplot(ax=axes[0], data=df, x='date', y='avg_goldstein', hue='country')
axes[0].set_title('Tono Promedio Goldstein en el Tiempo')
axes[0].set_ylabel('Goldstein')

sns.lineplot(ax=axes[1], data=df, x='date', y='n_conflict_events', hue='country')
axes[1].set_title('Eventos de Conflicto en el Tiempo')
axes[1].set_ylabel('Número de Eventos')

plt.tight_layout()
plt.show()
```

## 4. Correlación de Variables

```
numeric_cols = ['n_conflict_events', 'avg_goldstein', 'n_mentions', 
                'n_ucdp_events', 'total_fatalities', 'n_hotspots', 'avg_frp', 
                'n_news_articles', 'n_social_posts', 'avg_social_engagement']

plt.figure(figsize=(10, 8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```
